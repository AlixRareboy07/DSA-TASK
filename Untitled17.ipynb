```python
def merge(arr, a, b, c):
    x1 = b - a + 1
    x2 = c - b
 
    # Here, we create temp arrays
    L = [0] * (x1)
    R = [0] * (x2)
 
    # Copy the data to temp arrays L[] and R[]
    for i in range(0, x1):
        L[i] = arr[a + i]
 
    for j in range(0, x2):
        R[j] = arr[b + 1 + j]
 
    # Merge the temp arrays back into arr[l.....r]
    i = 0   
    j = 0    
    k = a    
 
    while i < x1 and j < x2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1
 
    # Copy the remaining elements of L[].
    while i < x1:
        arr[k] = L[i]
        i += 1
        k += 1
 
    # Copy the remaining elements of R[].
    while j < x2:
        arr[k] = R[j]
        j += 1
        k += 1
 
# (l) shows the left index and (r) shows the right one.
# sub-array of arr to be sorted
 
 
def mergeSort(arr, a, c):
    if a < c:
 
        # Same as (l+r)//2, but avoids overflow.
        b = a+(c-a)//2
 
        # Sort the halves
        mergeSort(arr, a, b)
        mergeSort(arr, b+1, c)
        merge(arr, a, b, c)
 
arr = [22, 21, 23, 15, 16, 17]
n = len(arr)
print("Given array is")
for i in range(n):
    print("%d" % arr[i],end=" ")
 
mergeSort(arr, 0, n-1)
print("\n\nSorted array is")
for i in range(n):
    print("%d" % arr[i],end=" ")
```

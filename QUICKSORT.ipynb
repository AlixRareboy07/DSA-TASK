```python
def partition(elements, start, end):
  
     index = start                      
     pivot  = elements[index]

     start=index + 1
     end=len(elements) - 1

     while start < end:
         while start < len(elements) and elements[start] <= pivot:
             start += 1
         while elements[end] > pivot:
             end -= 1
         if start < end:
             swap(start, end, elements)
        
     swap(index, end, elements)
     return end


def swap(a, b, element):
  if a != b:
     element[a],element[b]= element[b],element[a] 

def quick_sort(elements,start,end):    
    if start<end:
        recursion = partition(elements,start,end) 
        quick_sort(elements,start,recursion-1)
        quick_sort(elements,recursion+1,end)

elements = [66,13,5,55,14,79]
quick_sort(elements,0,len(elements)-1)
print(elements)
```
